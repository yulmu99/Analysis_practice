```python
pokemons = pd.read_csv('datas/pokemon.csv',index_col=['Pokemon']).squeeze()
pokemons
```

```python
#type 값이 2개이면 'multi' , 1개이면 'single' 반환하는 함수 만들기

def  multi_or_single(type) :
    if '/' in type:
        return 'multi'
    return 'single'
```

```python
pokemons.apply(multi_or_single)
```

```python
##DataFrame의 apply메서드

#apply(func, axis = )
#axis = 0 / index : 행마다 함수 적용
#axis = 1 / columns : 열마다 함수 적용
```

```python
df = pd.DataFrame([[4,9]] * 3 , columns=['A','B'])
df
```

```python
#행마다 함수 적용 / axis =0

df.apply(np.sum, axis =0 )
```

```python
#열마다 함수 적용 / axis=1

df.apply(np.sum, axis=1)
```

```python
df.apply(lambda x: [1,2], axis = 0)   #DataFrame 
```

```python
df.apply(lambda x: [1,2], axis = 1)   #Series
```

```python
#result_type ='expand'

#리스트 요소마다 하나의 값이 아닌 column으로 인식하도록 확장

df.apply(lambda x : [1,2], axis =1 , result_type = 'expand')
```

```python
#Series를 return
#result_type ='expand'와 비슷한 결과
#이때 Series의 index는 column label이 됨

df.apply(lambda x: pd.Series([1,2], index=['foo','bar']),axis =1)
```

```python
#result_type ='broadcast'

#함수로부터 반환되는게 리스트인지 스칼라인지 상관없이 axis 방향으로 브로드캐스팅
#column label은 유지
#기존의 shape과 함수 return값의 shape이 동일힐 때 broadcast 가능!   -> [3,2] 랑 [1,2] axis=1 가능

df.apply(lambda x : [1,2], axis =1,result_type  ='broadcast')
```

```python
#[3,2] 랑 [1,2,3]에서 2 != 3 으로 shape이 다름
#브로드캐스팅 불가

#df.apply(lambda x : [1,2,3], axis =1,result_type  ='broadcast')
```
