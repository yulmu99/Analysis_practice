### DataFrame Dropna() 메서드
###### row 단위로 동작
###### 기본적으로 행에 단 하나라도 결측지가 있으면 해당 행을 지움

```python
#how = 'any'  
#행 안에 값이 하나라도 결측치를 가지면 전체 행 제거

employees = pd.read_csv('datas/employees.csv',
                       parse_dates=['Start Date'])
employees.dropna()
```

```python
#how = 'all'
#행 안에 모든 값이 결측치를 가질 때만 행 제거

employees.dropna(how='all')
```

```python
#subset = []
#특정 column에 대해서만 결측치 제거

employees.loc[employees.Gender.isnull()]    #결측치 확인
employees.dropna(subset = ['Gender'])
```

```python
#Start Date , Salary column 둘 중 하나라고 결측치가 존재하면 제거

cond1= employees['Start Date'].isnull()
cond2= employees['Salary'].isnull()
print(f'cond1:{cond1.sum()}, cond2: {cond2.sum()}')

employees[cond1]
employees[cond2]

# |(vertical bar) - or 연산
employees[cond1|cond2]
```

```python
#thresh = n
#최소 n개 이상 값이 존재하는 행 추릴때
#thresh = 4 라면 최소 4개 이상 값이 있어야 보존 / 4개 미만인 경우에 행 제거
```

```python
employees[employees.notnull().sum(axis=1) >=4]

employees.dropna(thresh =4)
```

### DataFrame duplicated() 메서드

### DataFrame drop_duplicates() 메서드
##### 중복값 제거 
##### 기본적으로 행의 모든 값이 일치하는 행에 대해서 제거

```python
employees.drop_duplicates()    #열의 값이 모두 동일한 행이 없어서 제거되는 행이 없음
```

```python
#subset = []
#특정 column에 대한 중복 제거

employees.drop_duplicates(subset=['Team'])
```

```python
#keep = 'last'
#마지막에 발견된 값만 남기도 이전에 나오는 중복 제거

employees.drop_duplicates(subset=['Team'],keep = 'last')
```

#### DataFrame merge()
##### 공통 column을 기준으로 데이터 합치기

```python
#inner join 방식
#양쪽 DataFrame에 존재하는 키의 데이터만 보여주는 방식

pd.merge(df1,df2)
```

```python
##how='outer'
#outer join 방식
#키가 한 쪽에만 있어도 데이터 보여주는 방식

pd.merge(df1,df2,how='outer')
```

```python
##how='left'
#첫번째 인수 df를 기준으로 키 값을 모두 보여주는 방식

pd.merge(df1,df2,how='left')
```

```python
##how='right'
#두번째 인수 df를 기준으로 키 값을 모두 보여주는 방식

pd.merge(df1,df2,how='right')
```

#### DataFrame 그룹 분석

```python
groups = df2.groupby(df2.key1)
groups
```

```python
#A그룹과 B그룹 데이터의 합계

groups.sum()
```
