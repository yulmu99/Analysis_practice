```python
##DataFrame 고급 인덱싱
#DataFrame에서 특정 데이터만 골라내는 인덱싱

#loc인덱서 : label값 기반의 2차원 인덱싱
#iloc인덱서 : 순서를 나타내는 정수 기반의 2차원 인덱싱
```

```python
data = np.array(range(10,22)).reshape(3,4)
index = list('abc')
columns = list('ABCD')

df = pd.DataFrame(data,index=index,columns=columns)
```

```python
##loc인덱서
#.loc[row 인덱스, column 인덱스]

#row 단위 - 콤마(,)없으면 row기준으로만
    #row는 인덱스가 숫자, 문자(label) 두개 존재 
    #loc는 문자(lable)에 대해서 인덱싱 -> 슬라이싱 할 때 마지막 숫자도 포함

df.loc['a']
df.loc['b':'c']  #df['b':'c']
df.loc[['b','c']]   #b와 c row 각각 인덱싱 
```

```python
df.A > 15
```

```python
df.loc[df.A>15]   #True인 'c' row에 대해서 인덱싱
```

```python
def select_rows(df,num):
    return df.A > num

a = select_rows(df,10)
df.loc[a]
```

```python
##loc와 iloc 차이

#loc[] -> label(문자) 인덱싱 // 슬라이싱에서 마지막 포함
#iloc[] -> index(숫자) 인덱싱 // 마지막 포함 X

data = np.array(range(10,22)).reshape(3,4)
columns = list('ABCD')

df2 = pd.DataFrame(data,columns=columns)

df2.loc[1:2]

df2.iloc[1:2]
```

```python
#row,column 모두

df.loc['a','A']   #row index label이 'a'이고 column index label이 'A'인 자료
df.loc['b':,'A']
df.loc['a',:]
df.loc[['a','b'],['B','D']]

df.loc[df.A > 10, ['C','D']]  # column A의 값이 10이상인 row에 대해서 인덱싱, column은 'C'와 'D'에 대해서 인덱싱
```

```python
##iloc인덱서
#순서를 나타내는 정수 index만 
#다른 사항은 loc와 동일

df.iloc[2:3,1:3]
df.iloc[-1] = df.iloc[-1]*2    #인덱싱 자료 갱신
df
```
