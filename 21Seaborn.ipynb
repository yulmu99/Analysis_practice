## rugplot()
- 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내는 플롯

- 커널 밀도(kernel density): 커널이라는 함수를 겹치는 방법
    - 히스토그램보다 부드러운 형태의 분포 곡선

```python
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
```

```python
#Add a rug along both axes
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip")
```

### 색상별 분류
- hue = : 분류값 하나 더 추가

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue = "time")
sns.rugplot(data=tips, x="total_bill", y="tip", hue = "time")
```

### 플롯 크기 조절
- height = : 전체 그래프 크기 대비 차지 비율 ratio

```python
#height=.1 -> 전체 그래프의 10%

#Add a rug along both axes
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip", height=.1)  
```

### 투명도
- alpha = : 투명도 (1에 가까울수록 불투명)

```python
#lw = : line width

diamonds = sns.load_dataset("diamonds")
sns.scatterplot(data=diamonds, x="carat", y="price", s=5)
sns.rugplot(data=diamonds, x="carat", y="price", lw=1, alpha=.005)
```

#### Swarmplot()
- 데이터를 마커로 표현할 때 절대로 겹치지 않게 표현
- 정확한 데이터의 분포를 표현하기에 더 없이 좋은 그래프이지만, 아주 큰 스케일의 데이터를 표현하기에 필요한 너비가 너무 넓어져 사용하기에 적합하지 않음

```python
tips = sns.load_dataset('tips')
sns.swarmplot(data=tips,x='total_bill')
```

```python
# 각 카테고리 마다 분류

sns.swarmplot(data=tips , x='total_bill',y='day')
```

```python
#vertical 방향
#x값에 카테고리 자료형을 y값에 숫자형 데이터로 설정

sns.swarmplot(data=tips , y='total_bill',x='day')
```

```python
#hue인수
#변량 추가

sns.swarmplot(data=tips , x='total_bill',y='day',hue='sex')
```

```python
#dodge=True
#hue로 나눈 카테고리 값을 완전히 분리시켜 그래프로 표현

##strip plot과 다르게 swarm plot은 그래프의 크기가 데이터를 표현할만큼 충분히 크지 않으면 표현에 생략된 마커가 발생될 수 있습니다.

sns.swarmplot(data=tips , x='total_bill',y='day',hue='sex', dodge=True)
```

```python
#다양한 시각적 효과 적용 가능
#필요한 효과는 문서를 통해서 확인하고 사용하면 됨

sns.swarmplot(
    data=tips , x='total_bill',y='day',
    marker='x',linewidth=1)
```

### Pandas의 피벗테이블(pivot table)
- pivot table : 데이터 column 중에서 두 개의 column을 각각 row 인덱스, column 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것
- pivot_table(DataFrame,values=, index= , columns=, aggfunc='mean'(default))

```python
titanic = sns.load_dataset('titanic')
pd.pivot_table(titanic,'survived',index='class',columns='who')
```

```python
pd.pivot_table(titanic,'survived',index='class',columns='who',aggfunc=np.sum)
```

```python
#fill_value
#NaN값 처리

pd.pivot_table(titanic,values='survived',index='class',
               columns='deck',aggfunc=np.mean)
```

```python
pd.pivot_table(titanic,values='survived',index='class',
               columns='deck',aggfunc=np.mean,fill_value=0)
```

#### Heatmap

```python
ttn_cls_who = pd.pivot_table(titanic,'survived',index='class',columns='who')
sns.heatmap(ttn_cls_who,cmap='crest')
```

```python
#annot = True 
#각 cell마다 value 표기

#fmt = 
#value의 포맷 지정

#linewidth =
#cell 간의 구분
sns.heatmap(ttn_cls_who, cmap='mako',annot=True, fmt='.2f',linewidth=.5)
```
