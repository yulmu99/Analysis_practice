## rugplot() 
### 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 실제 데이터들의 위치를 보여준다

```python
tips = sns.load_dataset("tips")
tips
"""
total_bill : 총 지불 금액($)
tip : 팁($)
sex: 성별
smoker:흡연 여부
day:요일
time:식사 시간대
size:식사 인원
"""
```

```python
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
```

```python
#Add a rug along both axes
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip")
```

```python
# 색상을 나누어서 분류하기
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue = "time")
sns.rugplot(data=tips, x="total_bill", y="tip", hue = "time")
```

```python
#height=.1 -> 전체 그래프의 10%
```

```python
#Add a rug along both axes
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip", height=.1)
```

```python
#alpha -> 투명도 / lw => line width
```

```python
diamonds = sns.load_dataset("diamonds")
sns.scatterplot(data=diamonds, x="carat", y="price", s=5)
sns.rugplot(data=diamonds, x="carat", y="price", lw=1, alpha=.005)
```

## 히스토 그램 - displot()

```python
sns.displot(penguins, x="flipper_length_mm")
```

## 히스토그램 - displot() kde 로 표현하기

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde")
```

## 히스토그램 - displot() dcdf 로 표현하기

```python
#누적 분포 그래프 표현
sns.displot(penguins, x="flipper_length_mm", kind="ecdf")
```

## 히스토그램 - displot() 히스토그램 & kde 동시표현

```python
#kind = kde 비교 차이점
sns.displot(penguins, x="flipper_length_mm", kde=True)
```

## 히스토그램 - displot() 2개 값을 kde 표현

```python
# x,y축 모두 할당한 앞의 히스토그램을 kde로 표현 가능
sns.displot(penguins,
            x="flipper_length_mm",
            y="bill_length_mm",
            kind="kde",
            rug=True
)
```

## 히스토그램 - displot() hue로 subset 나누기

```python
# x,y축 모두 할당한 앞의 히스토그램을 kde로 표현 가능
sns.displot(penguins,
            x="flipper_length_mm",
            hue="species",
            kind = "kde"
)
```

## 히스토그램 - displot() col 나누기

```python
# displot()은 col 키워드 인수로 subset 한번더 분류 가능
sns.displot(penguins,
            x="flipper_length_mm",
            hue="species",
            col="sex",
            kind = "kde"
)
```

## 히스토그램 - displot() 그래프 크기 설정

```python
# hight와 aspect를 활용해서 그래프의 크기를 제어 할 수 있다.
sns.displot(penguins,
            y="flipper_length_mm",
            hue="sex",
            col="species",
            kind = "ecdf",
            height=4,
            aspect=.7
)
```

## 히스토그램 - displot() col로 그래프 나누기

```python
# axis label / title 지정 가능.
# axis label = set_axis_labels(xlabel.ylabel)
# title = set_titles() -> {col_var} / {col_name}
g = sns.displot(penguins,
                y="flipper_length_mm",
                hue="sex",
                col="species",
                kind = "kde",
                height=4,
                aspect=.7
)
g.set_axis_labels("Density (a.u.)", "Flipper length (mm)")
g.set_titles("{col_name} penguins")
```

# boxplot()

```python
#x값으로 series만을 전달. -> 수평하게 전달
df = sns.load_dataset("titanic")
sns.boxplot(x=df["age"])
```

# boxplot -> unique한 value의 개수만큼 얻기

```python
# class column의 타입은 카테고리이다.
#하나의 분류 값으로 추가하기 위해서 y키워드 인수로 설정
sns.boxplot(data=df, x = "age", y="class")
```

## -> column을 기준으로 그룹화

```python
# y와 hue에 전달된 column을 기준으로 그룹화
sns.boxplot(data=df, x = "age", y="class",hue="alive")
```

## -> 그래프의 순서 바꾸기 order()

```python
# order키워드 인수로 전달할 값을 리스트의 형태로 작성 -> ticklabel
sns.boxplot(data=df, x = "fare", y="alive")
sns.boxplot(data=df, x = "fare", y="alive",order=["yes","no"])
```

## -> 그래프의 순서 바꾸기 order()

```python
# box plot의 방향을 바꾸고 싶으면 orient키워드 인수에 h,v값으로 설정
sns.boxplot(data=df[["age","fare"]],orient="v")
```

# violinplot()

```python
#box plot과 kde(커널 밀도 함수)둘을 콤비네이션한 그래프 입니다.
# 생김새가 바이올린 같당
```

```python
# x키워드 인수만 값을 series로 할당하여 그렸다.
#기본적으로 수평 방향
df = sns.load_dataset("titanic")
sns.violinplot(x=df["age"])
```

```python
#class column을 활용하여 y축의 값으로 설정
#데이터 분포를 y축 기준으로 하여 분리해서 그릴 수 있다
sns.violinplot(data=df, x="age", y="class")
```

```python
# x키워드 인수에 카테고리 자료형을 설정 -> 방향 certical
sns.violinplot(data=df, x="class", y="age", hue="alive")
```

## split으로 반갈라 가능 대신 hue에 2개 값 갖는것만 가능

```python
#hue에 설정한 alive column의 값에 따라 좌측, 우측 각각 분포를 그린다.
#split키워드 인수에 True값을 전달하는 조건으로 hue에 level이 단 두개
sns.violinplot(data=df, x="deck", y="age", hue="alive",split=True)
```

```python
# 범위를 제한 할 수 있다. 나이의 경우 0미만의 범위에 그래프 표현 가능
sns.violinplot(data=df, x="age", y="alive")
```

```python
# 값의 표현 범위를 명확히 하기 위해 cut이라는 키워드 인수사용
sns.violinplot(data=df, x="age", y="alive",cut=0)
```

```python
# kde의 그래프 모양을 부드럽게 처리하다 보면 대역폭의 크기가 커짐
# bw 키워드 인수에 소수 값을 갖는 비율을 설정
sns.violinplot(data=df, x="age", y="alive",bw=.15)
```

```python
# 내부 데이터의 분포를 표현 inner = stick
sns.violinplot(data=df, x="age", y="embark_town",inner="stick")
```

```python
# scale에 따라 크기가 달라진다.(area,count,with)
sns.violinplot(data=df, x="age", y="embark_town",scale="count",cut=0)
```

# stripplot()
## -> 다른 축의 방향으로 랜덤하게 점을 찍어 겹피는 부분을 줄인다

```python
# 단 하나의 숫자형column을 할당 
#-> 각 점들이 다른축의 방향으로 랜덤하게 "jittered"단일 변량에 대한 분포
tips=sns.load_dataset("tips")
sns.stripplot(data=tips, x="total_bill")
```

```python
# y축의 방향에 카테고리 변량 추가
sns.stripplot(data=tips, x="total_bill",y="day")
```

```python
# x,y 바꾸기 -> 대조되게 vertical한 모양
sns.stripplot(data=tips, y="total_bill",x="day")
```

```python
# hue 키워드 인수를 추가해서 변량을 추가 할 수 있다.
sns.stripplot(data=tips, x="total_bill",y="day", hue="sex")
```

```python
# 이산형 숫자 타입의 데이터를 갖는 column도 지정 가능
sns.stripplot(data=tips, x="total_bill",y="day", hue="size")
```

```python
# dodge=True키워드 인수로 hue 로 나눈 카테고리 값을 완전히 분리
sns.stripplot(data=tips, x="total_bill",y="day", hue="sex", dodge=True)
```

# swarmplot()
## -> 절대로 겹치지 않게 표현 (아주 큰 데이터를 표현하기엔 적합하지 않음)

```python
# 숫자값을 갖는 단일 변량의 분포 표현
tips = sns.load_dataset("tips")
sns.swarmplot(data=tips, x="total_bill")
```

```python
# y값에 카테고리 자료형을 갖는 column을 추가
#요일별 지출액에 대한 분포를 나타냄
sns.swarmplot(data=tips, x="total_bill",y="day")
```

```python
# x값에 카테고리 자료형을 y값에 숫자형 데이터 설정 - 축 반대
sns.swarmplot(data=tips, y="total_bill",x ="day")
```

```python
# hue키워드 인수 추가 - 변량 추가
sns.swarmplot(data=tips, x="total_bill",y ="day", hue="sex")
```

```python
# dodge = True / hue로 나눈 값을 완전히 분리시킬 수 있다. 
#마커가 생략될 수있다.
sns.swarmplot(data=tips, x="total_bill",y ="day", hue="sex", dodge=True)
```

```python
# scatter() 메서드의 키워드 인수 작성으로 커스텀하게 적용 가능
sns.swarmplot(
    data=tips, x="total_bill",y ="day", 
    marker="x", linewidth=1,
)
```

# catplot()
## -> categorical plot = catplot

```python
#axes-level(row,col)까지 분류를 나눌 수 있다.
# kind키워드 인수를 사용하면 그래프 유형을 선택 할 수 있다.
```

```python
#기본은 jittered plot
df = sns.load_dataset("titanic")
sns.catplot(data=df, x="age", y="class")
```

```python
#kind 키워드 인수에 box라는 문자열 전달
sns.catplot(data=df, x="age",y="class",kind="box")
```

```python
#split으로 hue에는 두개의 값을 갖는 성별로 할당하여 두개로 나뉨
sns.catplot(
    data=df, x="age", y="class", hue="sex",
    kind="violin", bw=.25, cut=0, split=True,
)
```

```python
#kind = bar plot / axes가 2개인 것처럼 표현
#sex 분류에 따라 그래프가 col단위로 쪼개졌다.
#axes-level단위로 분류
sns.catplot(
    data=df, x="class", y="survived", col="sex",
    kind="bar", height=4, aspect=.6,
)
```
